```python
sla_targets = [[3,5,7,7],
              [5,5,7,7],
              [7,7,7,7],
              [7,7,7,7]]
```

```python
def det_country_index(country):
    if (country == 'metro manila'):
        return 0
    elif(country=='luzon'):
        return 1
    elif(country=='visayas'):
        return 2
    elif(country=='mindanao'):
        return 3
```

```python
def determine_sla_target(origin, destination):
    orig_ind = det_country_index(origin)
    dest_ind = det_country_index(destination)
    return sla_targets[orig_ind][dest_ind]
```

```python
determine_sla_target('metro manila', 'visayas')
```

```python
first['sla_target'] = np.vectorize(determine_sla_target)(first['origin'], first['destination'])
```

```python
orders['sla_target'] = np.vectorize(determine_sla_target)(orders['origin'], orders['destination'])
```

```python
def is_late_calc(first_del, second_del, sla):
    if second_del == None:
        if (first_del>sla):
            return 1
        else:
            return 0
    else:
        if((first_del<=sla) and (second_del<=3)):
            return 0
        else:
            return 1
```

```python
first['is_late'] = np.vectorize(is_late_calc)(first['1st_del_busday'], first['2nd_del_busday'], first['sla_target'])
```

```python
orders['is_late'] = np.vectorize(is_late_calc)(orders['1st_del_busday'], orders['2nd_del_busday'], orders['sla_target'])
```
